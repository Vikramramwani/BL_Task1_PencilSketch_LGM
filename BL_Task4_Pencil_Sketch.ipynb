```python
import cv2
```

```python
img=cv2.imread("Lion.JPG")
```

```python
cv2.imshow('Lion image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
plt.axis(False)
plt.show()
```

```python
plt.imshow(img[:,:,::-1])
plt.axis(False)
plt.show()
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis(False)
plt.show()
```

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.show()
```

```python
invert_img=cv2.bitwise_not(grey_img)
```

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
```

```python
invblur_img=cv2.bitwise_not(blur_img)
```

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
```

```python
cv2.imwrite('Lion Pencil sketch.png', sketch_img)
```

```python
cv2.imshow('Lion Pencil sketch',sketch_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Lion image', size=18)
plt.imshow(RGB_img)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Lion Pencil sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch_img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
